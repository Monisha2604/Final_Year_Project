```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('/content/drive/MyDrive/Final_project_processed_data.csv')
```

```python
X = data.drop("redshift",axis=1)   #Feature Matrix
y = data["redshift"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.tree import DecisionTreeRegressor
clf = DecisionTreeRegressor(max_depth=10, random_state=0)
```

```python
import time 

start=time.process_time()
clf.fit(X_train, y_train) 
stop=time.process_time()
execution_time = (stop - start)/60
print("Program Executed in minutes "+str(execution_time))
```

```python
Y_pred=clf.predict(X_test)
```

```python
from sklearn import metrics
print(metrics.r2_score(y_test,Y_pred))
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(111)
plt.title("Decision tree")
plt.scatter(y_test[:30000], Y_pred[:30000], s=0.30, color="purple")
ax1.set_xlabel("Test redshift")
ax1.set_ylabel("Predicted redshift")
ax1.plot([0,1], [0,1], '-r',color="black")
ax1.set_xlim(0, 1)
ax1.set_ylim(0, 1)

    
plt.show()
```

```python
from sklearn import metrics 
print(metrics.mean_absolute_error(y_test,Y_pred)) 
```

```python
print(metrics.mean_squared_error(y_test,Y_pred)) 
```

```python
print(np.sqrt(metrics.mean_squared_error(y_test,Y_pred)))
```
